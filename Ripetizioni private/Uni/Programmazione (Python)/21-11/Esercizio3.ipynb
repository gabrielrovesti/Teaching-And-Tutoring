```python
# Creazione lista dizionari articoli negozio
lista = [
    {'oggetto': 'Sci', 'prezzo': 800, 'quantita': 10},
    {'oggetto': 'Scarpone', 'prezzo': 20, 'quantita': 50},
    {'oggetto': 'Casco', 'prezzo': 50, 'quantita': 30},
]

# Lista degli articoli con le relative informazioni
print("Elementi nella lista")
for elemento in lista:
    print(f"Elemento: {elemento['oggetto']}, Prezzo: ${elemento['prezzo']}, Quantità disponibile: {elemento['quantita']}")

articolo_uno = {'Nome': 'Sci', 'prezzo': 20.00, 'quantita': 20}
print('\n'.join("{}: {}".format(k, v) for k, v in articolo_uno.items()))

for key, value in articolo_uno.items():
    print("{}: {}".format(key, value))

# while True:
#     oggetto_utente = input("\nInserisci il nome dell'articolo da acquistare: ")
#     quantita_utente = int(input("Inserisci la quantità da acquistare: "))
#
#     # Switch case - if abbreviato = varie possibilità
#     match(oggetto_utente):
#       case "Scarpone":
#         print("\nHai inserito uno scarpone")
#         break
#       case "Casco"
#
#
#     # il match (switch case) ti evita la scrittura estesa:
#
#     # if oggetto_utente == 'Sci' and 'Sci' in articolo_uno['Nome']
#     # or oggetto_utente == 'Scarpone' and 'Scarpone' in articolo_uno['Nome']
#     # or oggetto_utente == 'Casco' and 'Casco' in articolo_uno['Nome']
#
#     if oggetto_utente == 'sci' and 'Sci' in articolo_uno['Nome']:
#         print("Sono uno SCI")
#         break
#     else:
#         print(f"\nL'articolo inserito non è 'Sci' oppure 'Sci' non è presente in articolo_uno. Riprova.")

while True:
    oggetto_utente = input("\nInserisci il nome dell'articolo da acquistare: ")
    quantita_utente = int(input("Inserisci la quantità da acquistare: "))

    # Insieme di condizioni per oggetto_utente
    if oggetto_utente.lower() == 'scarpone':
        print("\nHai inserito uno scarpone")
    elif oggetto_utente.lower() == 'casco':
        print("\nHai inserito un casco")
    elif oggetto_utente.lower() == 'sci' and 'Sci' in articolo_uno['Nome']:
        print("\nHai inserito degli sci")
        break
    else:
        print("\nL'articolo inserito non corrisponde a uno degli articoli previsti. Riprova.")

    # Insieme di condizioni per articolo_uno['Nome']
    case_nome = articolo_uno['Nome']
    if case_nome.lower() == 'sci':
        print("Hai inserito degli sci")
        break
    elif case_nome.lower() == 'casco':
        print("Hai inserito un casco")
        break

    # Match (switch case - per semplificare le condizioni)
    match oggetto_utente:
      case "Scarpone":
          print("\n Hai inserito uno scarpone")
      case "Casco":
          print("\n Hai inserito un casco")
      case "Sci":
          print("\n Hai inserito degli sci")
      case _:
          print("\n Articolo non esistente")



 # Controllo sulle quantità (su tutta la lista)
    if quantita_utente < 0:
        print("Quantità negativa")
    else:
        item_found = False
        for elemento in lista:
            if elemento['oggetto'].lower() == oggetto_utente.lower():
                # Verifica se la quantità richiesta è disponibile
                if quantita_utente <= elemento['quantita']:
                    item_found = True

                    # Aggiornamento della quantità
                    elemento['quantita'] -= quantita_utente

                    # Calcolo del costo totale dell'ordine
                    total_cost = elemento['prezzo'] * quantita_utente

                    # Stampa messaggio di conferma
                    print(f"\nAcquisto confermato per {quantita_utente} {elemento['oggetto']}(s) a ${total_cost}.")

                    break

        if not item_found:
            print(f"\nMi dispiace, {quantita_utente} {oggetto_utente} non è disponibile.")

    # Calcola e stampa il totale delle entrate del negozio dopo gli acquisti
    total_store_receipts = sum(elemento['prezzo'] * (elemento['quantita'] - quantita_utente) for elemento in lista)
    print(f"\nTotale entrate del negozio dopo gli acquisti: ${total_store_receipts}")

    # Aggiungi un nuovo articolo alla lista e ristampa l'elenco aggiornato
    nuovo_articolo = {'oggetto': 'NuovoArticolo', 'prezzo': 30, 'quantita': 5}
    lista.append(nuovo_articolo)
    print("\nAggiunto nuovo articolo alla lista.")
    print("Elementi nella lista aggiornata")
    for elemento in lista:
        print(f"Elemento: {elemento['oggetto']}, Prezzo: ${elemento['prezzo']}, Quantità disponibile: {elemento['quantita']}")

    # Chiedi all'utente di inserire il nome di un articolo e rimuovilo dalla lista
    articolo_da_rimuovere = input("\nInserisci il nome dell'articolo da rimuovere dalla lista: ")
    lista = [elemento for elemento in lista if elemento['oggetto'].lower() != articolo_da_rimuovere.lower()]
    print(f"\nRimosso l'articolo {articolo_da_rimuovere} dalla lista.")

    # Stampa la lista finale degli articoli nel negozio
    print("\nLista finale degli articoli nel negozio")
    for elemento in lista:
        print(f"Elemento: {elemento['oggetto']}, Prezzo: ${elemento['prezzo']}, Quantità disponibile: {elemento['quantita']}")
```

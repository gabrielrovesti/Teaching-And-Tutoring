Una stringa è palindroma se può essere letta allo stesso modo da sinistra a destra e viceversa. Per esempio, "anna" è una stringa palindroma, mentre "casa" non lo è. 

Scrivi una funzione ricorsiva che prende in input una stringa e restituisce True se la stringa è palindroma, False altrimenti.

````python
# Esercizio sulla ricorsione per stringhe palindrome

def is_palindroma(stringa):
    if len(stringa) <= 1:
        return True
    else:
        return stringa[0] == stringa[-1] and is_palindroma(stringa[1:-1])

stringa = "giorgio"
print(is_palindroma(stringa)) 
````

Gioco Torre di Hanoi (spiegazione):
- tre bastoni ed n dischi
- i dischi sono di dimensioni diverse
- i dischi sono inizialmente impilati in ordine di dimensione su un bastone
- l'obiettivo è spostare tutti i dischi sul terzo bastone
- si può spostare un solo disco alla volta
- un disco più grande non può essere posto sopra un disco più piccolo

Generalmente, abbiamo A come bastone di partenza, B come bastone di appoggio e C come bastone di arrivo.

- Sposta n-1 dischi da A a C usando B
- Sposta il disco rimanente da A a B
- Sposta n-1 dischi da C a B usando A




````python
def hanoi(n, origine, destinazione, ausiliario):
    if n == 1:
        print("Sposta il disco 1 da", origine, "a", destinazione)
        return
    hanoi(n-1, origine, ausiliario, destinazione)
    print("Sposta il disco", n, "da", origine, "a", destinazione)
    hanoi(n-1, ausiliario, destinazione, origine)

n = 4
hanoi(n, "A", "B", "C")
````
